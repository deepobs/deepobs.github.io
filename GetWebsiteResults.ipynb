```python
import deepobs
```

```python
analyzer = deepobs.analyzer.analyze_utils.Analyzer("baselines_deepobs")
```

```python
problem = 'quadratic_deep'
runs = []
runs.append(analyzer.testproblems[problem].optimizers['MomentumOptimizer'].get_setting_most_runs().aggregate)
runs.append(analyzer.testproblems[problem].optimizers['AdamOptimizer'].get_setting_most_runs().aggregate)
runs.append(analyzer.testproblems[problem].optimizers['GradientDescentOptimizer'].get_setting_most_runs().aggregate)
```

```python
for run in runs:
    for idx, i in enumerate(['test_losses', 'train_losses']):
    #     print(i)
        if idx <0:
            print('{:.2%}'.format(run.output[i]['mean'][-1]), '{:.2%}'.format(max(run.output[i]['mean'])), end = ' ')
        else:
            print('{:.2f}'.format(run.output[i]['mean'][-1]), '{:.2f}'.format(min(run.output[i]['mean'])), end = ' ')
    print(format(run.output['speed'], '.1f'))
```
